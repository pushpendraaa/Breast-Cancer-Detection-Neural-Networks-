# Train Test Split

```python
X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
X_test.shape
```

```python
X_train.shape
```

# Data Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Creating the model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
model = Sequential()
model.add(Dense(units = 30,activation ='relu'))
model.add(Dense(units = 20, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600,validation_data=(X_test, y_test), verbose=1)
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

We have trained too much, probably we have to use early stopping to track the val_loss

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), verbose=1,callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict(X_test).round()
```

## Evalutation

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
